```python
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
with open("/Users/manaknarang/MANAK/Datasets/TCPD/datasets/bank/bank.json", "r") as read_file:
    data = json.load(read_file)
```

```python
vals = data['series'][0]['raw']
import pandas as pd
ind = np.arange(1,len(vals)+1)
dat = pd.DataFrame({'S.No.':ind,'Values':vals})
```

```python
dat["RMEAN2"] = dat.Values.rolling(2).mean()
dat["RMEAN5"] = dat.Values.rolling(5).mean()
dat["RMEAN20"] = dat.Values.rolling(20).mean()
dat["RMEAN50"] = dat.Values.rolling(50).mean()
dat["RMEAN100"] = dat.Values.rolling(100).mean()
```

```python
plt.plot(dat['S.No.'],dat['Values'],'o-',label='data')
plt.legend()
plt.show()
```

```python
plt.plot(dat["RMEAN5"],label = "5MA")
plt.plot(dat["RMEAN20"],label = "20MA")
plt.plot(dat["RMEAN50"],label = "50MA")
plt.plot(dat["RMEAN100"],label = "100MA")
plt.legend()
plt.show()
```
